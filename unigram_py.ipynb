```python
from collections import Counter

# Sample text
text = "I love NLP and I love Python."

# Step 1: Preprocess - lowercase and split into words
tokens = text.lower().split()

# Step 2: Count word frequencies (unigrams)
unigram_counts = Counter(tokens)

# Step 3: Calculate total number of words
total_tokens = sum(unigram_counts.values())

# Step 4: Calculate unigram probabilities
unigram_model = {word: count / total_tokens for word, count in unigram_counts.items()}

# Step 5: Print the unigram model
print("Unigram Model (word: probability):")
for word, prob in unigram_model.items():
    print(f"{word}: {prob:.4f}")
```
